```python
"""
IMDB Dataset Visualisation for LSTM Tutorial
Clean version with unified professional blue colour palette.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

# ---------------------------------------------------
# 1. Load the dataset
# ---------------------------------------------------

csv_path = r"C:\Users\sreep\OneDrive\Desktop\IMDB Dataset.csv"   # update if needed
df = pd.read_csv(csv_path)

print("Loaded:", df.shape)
print(df.head())

TEXT_COL = "review"
LABEL_COL = "sentiment"

# Convert labels to numeric
df[LABEL_COL] = df[LABEL_COL].map({"negative": 0, "positive": 1})

print("\nLabel counts:")
print(df[LABEL_COL].value_counts())

# ---------------------------------------------------
# 2. Text cleaning + tokenising
# ---------------------------------------------------

def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)            # remove HTML
    text = re.sub(r"[^a-zA-Z']", " ", text)       # keep letters and apostrophes
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_text"] = df[TEXT_COL].apply(clean_text)
df["tokens"] = df["clean_text"].apply(lambda t: t.split())
df["length"] = df["tokens"].apply(len)

print("\nLength stats:")
print(df["length"].describe())

# ---------------------------------------------------
# 3. Professional Blue Colour Palette
# ---------------------------------------------------

NAVY        = "#001f3f"
STEEL_BLUE  = "#4682B4"
SOFT_BLUE   = "#6CA6CD"
GREY        = "#7f7f7f"
LIGHT_GREY  = "#D3D3D3"

# ---------------------------------------------------
# 4. Plot 1  Sentiment Distribution
# ---------------------------------------------------

plt.figure(figsize=(6, 4))
counts = df[LABEL_COL].value_counts().sort_index()

plt.bar(
    ["Negative (0)", "Positive (1)"], 
    counts.values, 
    color=[NAVY, STEEL_BLUE],
    edgecolor="black"
)

plt.ylabel("Number of reviews")
plt.title("Sentiment Label Distribution")
plt.grid(axis="y", linestyle="--", color=LIGHT_GREY, alpha=0.6)
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 5. Plot 2  Histogram of Review Lengths
# ---------------------------------------------------

plt.figure(figsize=(8, 5))
plt.hist(
    df["length"],
    bins=50,
    color=SOFT_BLUE,
    edgecolor=GREY,
    alpha=0.9
)

plt.xlabel("Review length (tokens)")
plt.ylabel("Frequency")
plt.title("Distribution of Review Lengths")
plt.grid(axis="y", linestyle="--", color=LIGHT_GREY, alpha=0.6)
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 6. Plot 3  Boxplot by Sentiment
# ---------------------------------------------------

neg_lengths = df[df[LABEL_COL] == 0]["length"].values
pos_lengths = df[df[LABEL_COL] == 1]["length"].values

plt.figure(figsize=(7, 5))

box = plt.boxplot(
    [neg_lengths, pos_lengths],
    tick_labels=["Negative (0)", "Positive (1)"],
    patch_artist=True
)

# Apply colours
for patch, color in zip(box["boxes"], [NAVY, STEEL_BLUE]):
    patch.set_facecolor(color)
    patch.set_alpha(0.75)

plt.ylabel("Review length (tokens)")
plt.title("Review Length by Sentiment Class")
plt.grid(axis="y", linestyle="--", color=LIGHT_GREY, alpha=0.6)
plt.tight_layout()
plt.show()
```
